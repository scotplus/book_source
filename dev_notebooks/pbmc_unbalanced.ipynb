```python
import pickle
import scanpy as sc
y = pickle.load(open('./data/pbmc/final_rna.pkl', 'rb'))
X = pickle.load(open('./data/pbmc/final_atac.pkl', 'rb'))
adata = sc.read_h5ad('./data/pbmc/scRNA/adata.h5ad')
```

```python
import pandas as pd
X.columns = ["Region {0}".format(i + 1) for i in range(50)]
y.columns = ["Gene {0}".format(i + 1) for i in range(50)]
X
```

```python
ctypes = adata.obs['celltype']
```

```python
ctypes = pd.Series(ctypes).loc[X.index]
ctypes
```

```python
X = pd.concat((X, ctypes), axis=1)
y = pd.concat((y, ctypes), axis=1)
```

```python
[x for x in y.groupby("celltype")]
```

```python
import random
dfs = []
fracs = {}
for ctype, df in y.groupby("celltype"):
    print(ctype)
    fracs[df["celltype"][0]] = 0.5 + 0.05*random.choice(range(0, 11, 1))
    dfs.append(df.sample(frac=fracs[df["celltype"][0]], random_state=10))
```

```python
y_mod = pd.concat((dfs), axis=0)
```
