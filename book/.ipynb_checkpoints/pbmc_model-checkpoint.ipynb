```python
y_raw = adata.X
y_raw = y_raw.astype(np.float64)
y_raw
y_raw.shape
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=50)

y=pca.fit_transform(y_raw)

y.shape
```

```python
X_raw.shape
```

```python
scootr_sink = scoot.SCOOTR(nits_bcd=10, eval_bcd=1)
```

```python
# l2 distance
Dx_l2 = torch.cdist(torch.from_numpy(X_raw).to(device), torch.from_numpy(X_raw).to(device), p=2)**2
Dy_l2 = torch.cdist(torch.from_numpy(y).to(device), torch.from_numpy(y).to(device), p=2)**2

Dx_knn = scootr_sink.compute_graph_distances(X_raw, n_neighbors=25, mode='connectivity')
Dy_knn = scootr_sink.compute_graph_distances(y, n_neighbors=25, mode='connectivity')
```

```python
(pi_samp, pi_feat), _, _, _ = scootr_sink.solver_fgw(torch.from_numpy(Dx_knn).to(device),
        torch.from_numpy(Dy_knn).to(device),
        px=None,
        py=None,
        eps=5e-4,
        alpha=1,
        D=None,
        init_duals=None,
        verbose=True, 
        log=True,
        early_stopping_tol=1e-6)
```

```python
dump_mat(pi_samp, gen_mat_name('100423', 'GW', '1e-3', 'initpbmcsimplepca'))
```

```python
import umap
um = umap.UMAP(random_state=0)

originalX_um=um.fit_transform(aligned_X)
originaly_um=um.fit_transform(y)

#Visualization of the global geometry
fig, (ax1, ax2)= plt.subplots(1,2, figsize=(12, 5))
ax1.scatter(originalX_um[:,0], originalX_um[:,1], c="k", s=5)
ax1.set_title("ATAC Domain \n before Alignment")
ax2.scatter(originaly_um[:,0], originaly_um[:,1], c="b", s=5)
ax2.set_title("Gene Expression Domain \n before Alignment")
plt.show()
```
