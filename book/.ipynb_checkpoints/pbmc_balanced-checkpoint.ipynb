```python
import pickle
import numpy as np
import torch
import src.scootr as scoot
use_cuda = torch.cuda.is_available()
device = torch.device("cuda:0" if use_cuda else "cpu")
torch.backends.cudnn.benchmark=True
```

```python
y = pickle.load(open('./data/pbmc/final_rna.pkl', 'rb'))
X = pickle.load(open('./data/pbmc/final_atac.pkl', 'rb'))
X = X.to_numpy().astype('float64')
y = y.to_numpy().astype('float64')
```

```python
def gen_mat_name(date, method, eps, descrip):
    return '{0}_{1}_eps={2}_{3}'.format(date, method, eps, descrip)
def dump_mat(mat, name):
    pickle.dump(mat, open('./data/matrices/{0}.pkl'.format(name), 'wb'))
def load_mat(name):
    return pickle.load(open('./data/matrices/{0}.pkl'.format(name), 'rb'))
```

```python
eps_grid = np.logspace(-4, -1, 20)
```

```python
scootr_sink = scoot.SCOOTR(nits_bcd=10, eval_bcd=1)
Dx_knn = scootr_sink.compute_graph_distances(X, n_neighbors=25, mode='connectivity')
Dy_knn = scootr_sink.compute_graph_distances(y, n_neighbors=25, mode='connectivity')
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
for eps in np.flip(eps_grid):
    (pi_samp, _), _, _, _ = scootr_sink.solver_fgw(torch.from_numpy(Dx_knn).to(device),
        torch.from_numpy(Dy_knn).to(device),
        px=None,
        py=None,
        eps=1e-2,
        alpha=1,
        D=None,
        init_duals=None,
        verbose=False, 
        log=True,
        early_stopping_tol=1e-6)
    aligned_X = scootr_sink.get_barycentre(torch.from_numpy(y).to(device), torch.from_numpy(pi_samp.numpy()))
    fracs = scootr_sink.calc_domainAveraged_FOSCTTM(aligned_X.numpy(), y)
    legend_label="UGW alignment FOSCTTM \n average value: "+str(np.mean(fracs))
    plt.plot(np.arange(len(fracs)), np.sort(fracs), "b--", label=legend_label)
    plt.legend()
    plt.xlabel("Cells")
    plt.ylabel("Sorted FOSCTTM")
    plt.savefig('./data/pbmc/figs/initgridFGWFOSCTTMeps={0}.png'.format(eps))
    plt.close()
    pca=PCA(n_components=2)
    Xy_pca=pca.fit_transform(np.concatenate((aligned_X.numpy(), y), axis=0))
    X_pca=Xy_pca[0:2047,]
    y_pca=Xy_pca[2047:,]

    # Plot aligned domains, samples colored by domain identity:
    plt.scatter(y_pca[:,0], y_pca[:,1], c="b", s=5, label="Gene Expression")
    plt.scatter(X_pca[:,0], X_pca[:,1], c="r", s=5, label="ATAC")
    plt.legend()
    plt.title("Colored based on domains")
    plt.savefig('./data/pbmc/figs/initgridFGWPCAeps={0}.png'.format(eps))
    plt.close()
    
    pickle.dump(pi_samp, open('./data/matrices/initgridFGWeps={0}.pkl', 'wb'))
    
```
