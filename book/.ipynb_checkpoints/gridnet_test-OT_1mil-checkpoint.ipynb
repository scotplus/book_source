```python
import numpy as np
from scipy.sparse import csr_matrix

# Assuming atac_matrix is a sparse matrix in CSR format
# Transpose atac_matrix properly
atac_matrix_transposed = atac_matrix.transpose()

# Convert weighted_coupling to a sparse matrix in CSR format
weighted_coupling_sparse = csr_matrix(weighted_coupling)

# Perform the sparse matrix multiplication
result_sparse = weighted_coupling_sparse.dot(atac_matrix_transposed)
```

```python
def gene_info(x):
    # Extract gene names
#     print(x)
    g_name = list(filter(lambda x: 'gene_name' in x,  x.split(";")))[0].split("=")[1]
    g_type = list(filter(lambda x: 'gene_type' in x,  x.split(";")))[0].split("=")[1]
#     g_status = list(filter(lambda x: 'gene_status' in x,  x.split(";")))[0].split("=")[1]
    g_leve = int(list(filter(lambda x: 'level' in x,  x.split(";")))[0].split("=")[1])
    return (g_name, g_type, g_leve)
```

```python
gene_list = rna_adata.var['features']

gencode = pd.read_table("./gencode.v38.annotation.gff3.gz", comment="#",
                    sep = "\t", names = ['seqname', 'source', 'feature', 'start' , 'end', 'score', 'strand', 'frame', 'attribute'])

gencode_genes = gencode[(gencode.feature == "gene")][['seqname', 'start', 'end', 'attribute']].copy().reset_index().drop('index', axis=1) # Extract genes

gencode_genes["gene_name"], gencode_genes["gene_type"], gencode_genes["gene_level"] = zip(*gencode_genes.attribute.apply(lambda x: gene_info(x)))
```

```python
%%time
chunks = np.array_split(candidates_df, len(candidates_df) // 4096 + 1)


from scipy.sparse import issparse
X = atac_adata.X.toarray() if issparse(atac_adata.X) else atac_adata.X
Y = rna_adata.X.toarray() if issparse(rna_adata.X) else rna_adata.X
results_df = pd.DataFrame()
print("length: ",len(chunks))
for i in range(0,len(chunks)):
    print(i)
    X_feature_names = chunks[i]['atac_id'].unique()
    Y_feature_names = chunks[i]['gene'].unique()
    candidate_XY_pairs = [(x,y) for x,y in chunks[i][['atac_id','gene']].values]

    results_df = pd.concat([results_df,gridnet.run_gridnet(X,Y,X_feature_names,Y_feature_names,

                                     candidate_XY_pairs,
                                     dag_adjacency_matrix,
                                     device ='cuda:0',
                                    shuffle=True,
#                                     save_dir="./gridnet_multiomic_original_chunks_{}".format(i),
                                    max_epochs=200,
                                    batch_size = 4096,
                                    verbose = False)],
             axis = 0)
```
